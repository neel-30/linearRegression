```python
q1=df.Anual_Salary.quantile(.25)
q3=df.Anual_Salary.quantile(.75)
iqr=q3-q1
ub=q3+1.5*iqr
lb=q1-1.5*iqr
df=df[(df['Anual_Salary']>=lb )& (df['Anual_Salary']<=ub)]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x=df.loc[:,['age','bmi','children','Claim_Amount','past_consultations','num_of_steps','Hospital_expenditure','NUmber_of_past_hospitalizations','Anual_Salary']]
y=df.loc[:,'charges']          
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size =0.80,random_state=0) #splitting the data
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train) #training
```

```python
predict=model.predict(x_test)
```

```python
result=pd.DataFrame(columns=['actual','predicted'])
result['actual']=y_test
result['predicted']=predict
result
```

```python
from sklearn.metrics import *
r2_score(y_test,predict)*100        #accuracy is 95%
```

```python
sns.regplot(x=y_test,y=predict)
plt.xlabel('Actual Data')
plt.ylabel('Predicted Data')
```
